```python
import pandas as pd 
import numpy as np 
import pickle 
from scipy.stats import boxcox
from scipy.special import inv_boxcox
from statsmodels.formula.api import ols
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_parquet('/data/u_AA/throughput_opt/tiering_cells.parquet')

for feature in df.select_dtypes(['float64','int64']).columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df[feature], y=df['DL_THROUGHPUT'])
    plt.title(f'Scatter Plot: {feature} vs Target')
    plt.xlabel(feature)
    plt.ylabel('Target')
    plt.show()
```

```python
for feature in df.select_dtypes(['float64','int64']).columns:
    transformed_target,_= boxcox(df['DL_THROUGHPUT'])
    transformed_feature,_= boxcox(df[feature])
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=transformed_feature, y=transformed_target,alpha=0.5)
    plt.title(f'Scatter Plot: {feature} vs Target')
    plt.xlabel(feature)
    plt.ylabel('Target')
    plt.show()
```
